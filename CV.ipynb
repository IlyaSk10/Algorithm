```
import youtube_dl

# ф-ия скачивания видео
def download_video(name_of_video): 
  for video in name_of_video:
    ydl_opts = {'outtmpl': video,'format':'134',} # имя файла video, с разрешением 360p.

    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        ydl.download(['https://www.youtube.com/watch?v='+video])
  return print('скачано')
```

```
# функция извлечения времени, чтобы понять на какой секунде следует захватить кадр с видео
def extract_times_from_video(name_of_video):
  video_times=dict()
  for video in name_of_video:
    times=list()
    classes=list(daly['annot'][video]['annot'].keys())
  
    for i in range(len(classes)):
      length=len(daly['annot'][video]['annot'][classes[i]])
      for t in range(length):
        number_keyframes=len(daly['annot'][video]['annot'][classes[i]][t]['keyframes'])
        times.append([daly['annot'][video]['annot'][classes[i]][t]['keyframes'][f]['time'] for f in range(number_keyframes)])
    video_times.update({video:times})
  return video_times
```

```
times=extract_times_from_video(video_names[:2])
```

```
import cv2
import os
import re

# ф-ия захвата кадра с видео по меткам времени из times
def capture_frames(name_of_video,times):
  for name in name_of_video:
    new_dir=str(re.findall(r'(\w+)\.\w+',name)[0])
    os.mkdir(new_dir)  # создаем папку с номером видео,в этой папке будут храниться все кадры с опред-го видео
  
    for time in times[name]:
      for frame in time:
        vidcap = cv2.VideoCapture(name)
        vidcap.set(cv2.CAP_PROP_POS_MSEC,frame*1000)      # установить позицию во времени (в секундах)
        success,image = vidcap.read()
        if success:
          os.chdir('/content/'+new_dir)  # переход в папку с названием видео  
          cv2.imwrite("%s.jpg"%(frame), image)     # сохранить кадр в N-ую секунду 
          cv2.waitKey()
          os.chdir('/content/') # возвращаемся обратно в /content/
  return print('возвращены кадры')    
```

```
# ф-ия конвертация в RGB формат
def convert2rgb(name_of_times):
  img = cv2.imread(str(name_of_times)+'.jpg')
  img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  return img_rgb
```

Далее необходимо подготовить и нарисовать bbox и label на картинках. Делается это с помощью cv2.rectangle(). Координаты bbox и label есть в разметке у каждого frame
Эти картинки с bbox и label и есть так называемые правильные ответы классификатора.
То есть на вход подается original image, далее делается прогноз и на выходе получаем картинку с label(классификация), определяются границы bbox(регрессия) и прогноз сравнивается с правильным ответом.
В процессе обучения веса настраиваются с учетом минимизации ошибки.

Далее рассчитывается predict как процент правильных ответов классификатора и доля пересечений предсказанных bounding box с истинным положением (метрика MAP)

```
# подсчет распределения классов
import collections
c = collections.Counter()
vid=list(daly['annot'].keys())
for number in vid:
  classes=daly['annot'][number]['annot'].keys()
  for key in classes:
    c[key]+=len(daly['annot'][number]['annot'][key])
```

```
# график распределения классов
import matplotlib.pyplot as plt
plt.bar(c.keys(), c.values())
plt.xlabel('Class')
plt.ylabel('Number of instances')
plt.xticks(rotation=75)
```

В задаче классификации по графику видно, что классы не сбалансированы, а значит метрика accuracy_score не подходит. Лучше подойдет precision, recall, confusion matrix. В случае многоклассовой классификации можно использовать микро- или макро- усреднение.
